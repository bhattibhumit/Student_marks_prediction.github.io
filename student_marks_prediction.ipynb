```python
# Problem

# 1.how many hours need to do the study to get 99% marks ? 
# 2.how many hours need to do the study to pass the exam ?
# 3.If i will do study x(4) hours per day so how much marks i will get

```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("student_info.csv")
print(df.shape)
```

```python
print(df.head())
print(df.tail())
```

```python
# Data Visual 

sns.lmplot(x='study_hours',y='student_marks',data=df)
```

```python
# we can also matlabplot 
plt.figure(figsize=(7,7))
plt.scatter(x=df.study_hours,y=df.student_marks)
plt.xlabel("Study Hours")
plt.ylabel("Student Marks")
plt.title("Student Marks Prediction")
```

```python
# data Cleaning

df.isnull().sum()
```

```python
# fill with Mean 

df2 = df.fillna(df.mean())

print(df2.head())

df2.isnull().sum()
```

```python
# split data

x = df2.drop("student_marks",axis=1)
y = df2.drop("study_hours",axis=1)

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=51)
print("shape of X_train :", X_train.shape)
print("shape of Y_train :", X_test.shape)
print("shape of X_train :", y_train.shape)
print("shape of X_train :", y_test.shape)
```

```python
# select the model and train it

from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
print(lr.coef_) # it slop value(m)
print(lr.intercept_) # it is constant value(c)
```

```python
lr.predict([[4]])[0][0].round(2)
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
pd.DataFrame(np.c_[X_test,y_test,y_pred] , columns = ["Study_hours","Student_marks","Predicated_marks"])
```

```python
plt.scatter(X_test,y_test)
plt.plot(X_train,lr.predict(X_train),color="r")
plt.xlabel("Study Hours")
plt.ylabel("Student Marks")
plt.title("Student Marks Prediction")

```

```python
# to find accuracy
lr.score(X_test,y_test)
```

```python
# Save Your ML model

import joblib as j
j.dump(lr,"Student_Marks_Predication_model.pkl")
```

```python
model = j.load("Student_Marks_Predication_model.pkl")
model.predict([[4]])
```
